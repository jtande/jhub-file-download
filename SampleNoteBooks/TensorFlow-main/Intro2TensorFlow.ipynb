## What is TensorFlow?

TensorFlow is **Google's second machine learning framework** for designing, building, and training deep learning models.


## Why to use TensorFlow?

The TensorFlow library may be used to perform numerical computations, which may not appear to be particularly noteworthy in and of itself, but these computations are carried out using data flow graphs.

## Where in the real world we use TensorFLow?

TensorFlow is widely used across the world by **Data scientists**. <br/>
This is an open-source deep learning and machine learning library that has applications in text-based applications, image recognition, voice search, and many other areas. TensorFlow is used for image identification in DeepFace, **Facebook's image recognition** technology. **Apple's Siri** uses it for **voice recognition**.

## How to use Tensor flow?
